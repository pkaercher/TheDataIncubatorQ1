#### Question 3
We can use the FDNY dataset to investigate at what time of the day people cook most. Compute what proportion of all incidents are cooking fires for every hour of the day by normalizing the number of cooking fires in a given hour by the total number of incidents that occured in that hour. Find the hour of the day that has the highest proportion of cooking fires and submit that proportion of cooking fires. A cooking fire is an incident for which 'INCIDENT_TYPE_DESC' is '113 - Cooking fire, confined to container'. Note: round incident times down. For example, if an incident occured at 22:55 it occured in hour 22.

```python
fdny.insert(3,'INCIDENT_TIME',pd.to_datetime(fdny['INCIDENT_DATE_TIME']))
fdny['INCIDENT_TIME'] = fdny['INCIDENT_TIME'].dt.strftime('%Y-%m-%d %H:%M:%s')
fdny['INCIDENT_TIME'] = fdny['INCIDENT_TIME'].str[10:13]

fdny
```

```python
cooking_fire = fdny[fdny['INCIDENT_TYPE_DESC']=='113 - Cooking fire, confined to container']
other = fdny[fdny['INCIDENT_TYPE_DESC']!='113 - Cooking fire, confined to container']
cf_counts = cooking_fire[['IM_INCIDENT_KEY','INCIDENT_TYPE_DESC','INCIDENT_TIME']].groupby(['INCIDENT_TYPE_DESC','INCIDENT_TIME']).agg(['count'])
other_counts = other[['IM_INCIDENT_KEY','INCIDENT_TIME']].groupby(['INCIDENT_TIME']).agg(['count'])
```

```python
new_df = pd.DataFrame({'INCIDENT_TIME':other_counts.index,\
                       'COOKING_FIRE_COUNTS':list(cf_counts[('IM_INCIDENT_KEY','count')]),
                       'OTHER_COUNTS':list(other_counts[('IM_INCIDENT_KEY','count')])})

new_df['NORMALIZE_CF_COUNT'] = new_df['COOKING_FIRE_COUNTS']/new_df['OTHER_COUNTS']
# new_df[new_df['NORMALIZE_CF_COUNT'] == new_df['NORMALIZE_CF_COUNT'].max()]
new_df['NORMALIZE_CF_COUNT'].max()
```

#### Question 4
What is the ratio of the average number of units that arrive to a scene of an incident classified as '111 - Building fire' to the number that arrive for '651 - Smoke scare, odor of smoke'?

```python
building_fire = fdny[fdny['INCIDENT_TYPE_DESC']=='111 - Building fire']
smoke_scare = fdny[fdny['INCIDENT_TYPE_DESC']=='651 - Smoke scare, odor of smoke']

ave_units_bf = building_fire['UNITS_ONSCENE'].mean()
ave_units_ss = smoke_scare['UNITS_ONSCENE'].mean()

print(ave_units_bf/ave_units_ss)
```

#### Question 5
Check the distribution of the number of minutes it takes between the time a '111 - Building fire' incident has been logged into the Computer Aided Dispatch system and the time at which the first unit arrives on scene. What is the third quartile of that distribution. Note: the number of minutes can be fractional (ie, do not round).

```python
import numpy as np

building_fire = fdny[fdny['INCIDENT_TYPE_DESC']=='111 - Building fire']
building_fire.dropna(subset=['INCIDENT_DATE_TIME','ARRIVAL_DATE_TIME'], inplace=True)

# convert time strings into datetimes to compare and calculate difference between
for col in ['INCIDENT_DATE_TIME','ARRIVAL_DATE_TIME']:
    building_fire[col] = pd.to_datetime(building_fire[col])

# throw out arrival time that are later than incident times
building_fire = building_fire[building_fire['ARRIVAL_DATE_TIME']>=building_fire['INCIDENT_DATE_TIME']]

building_fire.insert(5,'RESPONSE_TIME',building_fire['ARRIVAL_DATE_TIME'] - building_fire['INCIDENT_DATE_TIME'])

# convert from time delta in nanoseconds to minutes as a float
building_fire['RESPONSE_TIME_MINS'] = building_fire['RESPONSE_TIME'].values.astype(np.int64)*(10**(-9))/60

median_resp_time = np.median(building_fire['RESPONSE_TIME_MINS'])
upper_half = [x for x in list(building_fire['RESPONSE_TIME_MINS']) if x >= median_resp_time]
third_quartile = np.median(upper_half)

third_quartile
```
